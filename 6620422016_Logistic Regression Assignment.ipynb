# 2.Clean data


```python
# Check for missing data in the dataset
missing_data = df_smartphone.isnull().sum()

# Display columns with missing data (if any)
missing_data[missing_data > 0]
```

```python
# Check for duplicate entries
duplicates =  df_smartphone.duplicated()

# Remove duplicate entries, if any
data_cleaned =  df_smartphone.drop_duplicates()

# Check the number of duplicates removed
duplicates_removed = duplicates.sum()

duplicates_removed
```

```python
# Renaming the columns into English for easier processing
translated_columns = {
    'Timestamp': 'Timestamp',
    'เพศของผู้ตอบแบบสอบถาม': 'Gender',
    'ช่วงอายุของผู้ตอบแบบสอบถาม': 'Age Group',
    'อาชีพของผู้ตอบแบบสอบถาม': 'Occupation',
    'ช่วงเงินเดือนของผู้ตอบแบบสอบถาม': 'Salary Range',
    'ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด': 'Current Smartphone OS',
    'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [เล่นเกม]': 'Most Used For [Gaming]',
    'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [เล่นโซเชี่ยลมีเดีย]': 'Most Used For [Social Media]',
    'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [ดูหนัง ฟังเพลง]': 'Most Used For [Movies/Music]',
    'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [ถ่ายรูป ถ่ายวิดีโอ]': 'Most Used For [Photos/Videos]',
    'หากท่านซื้อสมาร์ทโฟนเครื่องใหม่ จะใช้เลือกซื้อสมาร์ทโฟนแบรนด์ใด': 'New Smartphone Brand Choice',
    'จากข้อที่แล้ว เนื่องจากสาเหตุใด': 'Reason for Brand Choice'
}

# Apply the translation to the column names
data_cleaned.columns = data_cleaned.columns.to_series().map(translated_columns)

# Display the first few rows to confirm the column renaming
data_cleaned.head()
```

# 3.Transform to Lable Encoder

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
label_encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = [
    'Gender',
    'Age Group',
    'Occupation',
    'Salary Range',
    'Current Smartphone OS',
    'Most Used For [Gaming]',
    'Most Used For [Social Media]',
    'Most Used For [Movies/Music]',
    'Most Used For [Photos/Videos]',
    'New Smartphone Brand Choice'
]

# Apply label encoding to the selected columns
encoded_data = data_cleaned.copy()
for column in columns_to_encode:
    encoded_data[column] = label_encoder.fit_transform(data_cleaned[column])


encoded_data.head()
```

# 4.Split data to Train data 70% and Test data 30%

```python
from sklearn.model_selection import train_test_split

# Splitting the data into train and test sets (70% train, 30% test)
train_data, test_data = train_test_split(encoded_data, test_size=0.3, random_state=42)

train_data
test_data
```

# 5.Train Model Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create a binary column for whether the chosen brand is iPhone or not
train_data['target'] = (train_data['New Smartphone Brand Choice'] == 1).astype(int)
test_data['target'] = (test_data['New Smartphone Brand Choice'] == 1).astype(int)

# Drop unnecessary columns for training (like the target and non-numeric columns)
X_train = train_data.drop(columns=['Timestamp', 'Reason for Brand Choice', 'target', 'New Smartphone Brand Choice'])
X_test = test_data.drop(columns=['Timestamp', 'Reason for Brand Choice', 'target', 'New Smartphone Brand Choice'])

y_train = train_data['target']
y_test = test_data['target']

# Initialize the logistic regression model
log_reg_model = LogisticRegression(max_iter=1000)

# Fit the model to the training data
log_reg_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = log_reg_model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

accuracy
```
